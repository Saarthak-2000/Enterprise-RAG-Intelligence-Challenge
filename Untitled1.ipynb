```python
import os
import json
import pandas as pd

from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_community.vectorstores import FAISS
from langchain_core.documents import Document
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_ollama import OllamaLLM
```

```python
os.makedirs("data/csv", exist_ok=True)
os.makedirs("data/logs", exist_ok=True)
os.makedirs("data/reports", exist_ok=True)
```

```python
employees = pd.DataFrame({
    "employee_id": [1, 2, 3],
    "name": ["John", "Alice", "Bob"],
    "department": ["finance", "security", "general"],
    "salary": [90000, 110000, 60000]
})

employees.to_csv("data/csv/employees.csv", index=False)

employees.head()
```

```python
logs = [
    {
        "timestamp": "2025-08-01",
        "severity": "critical",
        "service": "payment-api",
        "message": "Unauthorized login attempt"
    },
    {
        "timestamp": "2025-08-02",
        "severity": "medium",
        "service": "gateway-api",
        "message": "Timeout issue detected"
    }
]

with open("data/logs/security_logs.json", "w") as f:
    json.dump(logs, f)

print("Security logs created")
```

```python
finance_report = """
Q4 Finance Report

Infrastructure spending exceeded allocated budget.
Authentication vulnerabilities were identified during audit.
Operational costs increased by 18 percent.
"""

with open("data/reports/finance_report.txt", "w") as f:
    f.write(finance_report)

print("Finance report created")
```

```python
csv_docs = []

for file in os.listdir("data/csv"):

    if file.endswith(".csv"):

        df = pd.read_csv(os.path.join("data/csv", file))

        for _, row in df.iterrows():

            csv_docs.append(
                Document(
                    page_content=row.to_json(),
                    metadata={
                        "source": file,
                        "department": row.get("department", "general"),
                        "type": "csv"
                    }
                )
            )

print("CSV docs loaded:", len(csv_docs))
```

```python
json_docs = []

for file in os.listdir("data/logs"):

    if file.endswith(".json"):

        with open(os.path.join("data/logs", file), "r") as f:

            data = json.load(f)

            for log in data:

                json_docs.append(
                    Document(
                        page_content=json.dumps(log),
                        metadata={
                            "source": file,
                            "department": "security",
                            "type": "json_log"
                        }
                    )
                )

print("JSON docs loaded:", len(json_docs))
```

```python
report_docs = []

for file in os.listdir("data/reports"):

    path = os.path.join("data/reports", file)

    with open(path, "r") as f:

        text = f.read()

        report_docs.append(
            Document(
                page_content=text,
                metadata={
                    "source": file,
                    "department": "finance",
                    "type": "report"
                }
            )
        )

print("Report docs loaded:", len(report_docs))
```

```python
all_docs = csv_docs + json_docs + report_docs

print("Total documents:", len(all_docs))
```

```python
splitter = RecursiveCharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=100
)

chunked_docs = splitter.split_documents(all_docs)

print("Chunked documents:", len(chunked_docs))
```

```python
embedding_model = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)

print("Embedding model loaded")
```

```python
vector_db = FAISS.from_documents(
    chunked_docs,
    embedding_model
)

vector_db.save_local("vector_store")

print("FAISS vector database created successfully")
```

```python
USER_ROLES = {

    "alice": {
        "role": "FinanceAdmin",
        "departments": ["finance"]
    },

    "bob": {
        "role": "SecurityAnalyst",
        "departments": ["security"]
    },

    "charlie": {
        "role": "Employee",
        "departments": ["general"]
    }
}
```

```python
ROLE_LEVELS = {
    "Employee": 1,
    "FinanceAdmin": 3,
    "SecurityAnalyst": 3
}

DOC_LEVELS = {
    "general": 1,
    "finance": 3,
    "security": 3
}


def is_authorized(username, metadata):

    user = USER_ROLES.get(username)

    if not user:
        return False

    user_level = ROLE_LEVELS[user["role"]]

    department = metadata.get("department", "general")

    required_level = DOC_LEVELS[department]

    if user_level < required_level:
        return False

    if department not in user["departments"] and department != "general":
        return False

    return True
```

```python
def retrieve_documents(query, username, k=5):

    docs = vector_db.similarity_search(query, k=k)

    authorized_docs = []

    for doc in docs:

        if is_authorized(username, doc.metadata):

            authorized_docs.append(doc)

    return authorized_docs
```

```python
query = "finance budget risks"

results = retrieve_documents(query, "alice")

for doc in results:

    print("SOURCE:", doc.metadata["source"])
    print(doc.page_content)
    print("-" * 50)
```

```python
llm = OllamaLLM(model="llama3")

print("Llama3 loaded successfully")
```

```python
SYSTEM_PROMPT = """
You are a secure enterprise AI assistant.

Answer ONLY using the provided context.

Do not hallucinate.

Always provide grounded answers.
"""


def generate_answer(query, docs):

    context = "\n\n".join([
        d.page_content for d in docs
    ])

    citations = list(set([
        d.metadata["source"] for d in docs
    ]))

    prompt = f"""
{SYSTEM_PROMPT}

Context:
{context}

Question:
{query}
"""

    answer = llm.invoke(prompt)

    return {
        "answer": answer,
        "citations": citations,
        "confidence": "High" if len(docs) >= 2 else "Medium"
    }
```

```python
from langchain_ollama import OllamaLLM

# Use the model you actually downloaded
llm = OllamaLLM(model="llama3.2") 
```

```python
query = "What finance risks were detected?"

retrieved_docs = retrieve_documents(query, "alice")

response = generate_answer(query, retrieved_docs)

response
```

```python
print("ANSWER:")
print(response["answer"])

print("\nCITATIONS:")

for c in response["citations"]:
    print("-", c)

print("\nCONFIDENCE:")
print(response["confidence"])
```
