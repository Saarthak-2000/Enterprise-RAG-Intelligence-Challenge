```python
import os
import json
import pandas as pd

from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_community.document_loaders import PyPDFLoader
from langchain_community.vectorstores import FAISS
from langchain_openai import OpenAIEmbeddings
from langchain_core.documents import Document

from openai import OpenAI
```

```python
os.environ["OPENAI_API_KEY"] = "YOUR_OPENAI_API_KEY"

client = OpenAI()
```

```python
os.makedirs("data/pdfs", exist_ok=True)
os.makedirs("data/csv", exist_ok=True)
os.makedirs("data/logs", exist_ok=True)
```

```python
employees = pd.DataFrame({
    "employee_id": [1, 2, 3],
    "name": ["John", "Alice", "Bob"],
    "department": ["finance", "security", "general"],
    "salary": [90000, 110000, 60000]
})

employees.to_csv("data/csv/employees.csv", index=False)

employees.head()
```

```python
logs = [
    {
        "timestamp": "2025-08-01",
        "severity": "critical",
        "service": "payment-api",
        "message": "Unauthorized login attempt"
    },
    {
        "timestamp": "2025-08-02",
        "severity": "medium",
        "service": "gateway-api",
        "message": "Timeout issue detected"
    }
]

with open("data/logs/security_logs.json", "w") as f:
    json.dump(logs, f)

print("JSON logs created")
```

```python
csv_docs = []

for file in os.listdir("data/csv"):
    if file.endswith(".csv"):
        df = pd.read_csv(os.path.join("data/csv", file))

        for _, row in df.iterrows():
            csv_docs.append(
                Document(
                    page_content=row.to_json(),
                    metadata={
                        "source": file,
                        "type": "csv",
                        "department": row.get("department", "general")
                    }
                )
            )

print(f"CSV Documents Loaded: {len(csv_docs)}")
```

```python
json_docs = []

for file in os.listdir("data/logs"):
    if file.endswith(".json"):

        with open(os.path.join("data/logs", file), "r") as f:
            data = json.load(f)

            for log in data:
                json_docs.append(
                    Document(
                        page_content=json.dumps(log),
                        metadata={
                            "source": file,
                            "type": "json_log",
                            "department": "security"
                        }
                    )
                )

print(f"JSON Documents Loaded: {len(json_docs)}")
```

```python
pdf_docs = []

for file in os.listdir("data/pdfs"):
    path = os.path.join("data/pdfs", file)

    with open(path, "r") as f:
        text = f.read()

        pdf_docs.append(
            Document(
                page_content=text,
                metadata={
                    "source": file,
                    "type": "report",
                    "department": "finance"
                }
            )
        )

print(f"PDF/Text Documents Loaded: {len(pdf_docs)}")
```

```python
all_docs = csv_docs + json_docs + pdf_docs

print(f"Total Documents: {len(all_docs)}")
```

```python
splitter = RecursiveCharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=100
)

chunked_docs = splitter.split_documents(all_docs)

print(f"Chunked Documents: {len(chunked_docs)}")
```

```python
embedding_model = OpenAIEmbeddings()
```

```python
import os
from langchain_openai import OpenAIEmbeddings

os.environ["OPENAI_API_KEY"] = "sk-your-real-key"

embedding_model = OpenAIEmbeddings()
```

```python
vector_db = FAISS.from_documents(
    chunked_docs,
    embedding_model
)

vector_db.save_local("vector_store")

print("Vector database created successfully")
```

```python
USER_ROLES = {
    "alice": {
        "role": "FinanceAdmin",
        "departments": ["finance"]
    },
    "bob": {
        "role": "SecurityAnalyst",
        "departments": ["security"]
    },
    "charlie": {
        "role": "Employee",
        "departments": ["general"]
    }
}
```

```python
ROLE_ACCESS = {
    "Employee": 1,
    "FinanceAdmin": 3,
    "SecurityAnalyst": 3
}

DOC_ACCESS = {
    "general": 1,
    "finance": 3,
    "security": 3
}


def is_authorized(username, metadata):

    user = USER_ROLES.get(username)

    if not user:
        return False

    user_level = ROLE_ACCESS[user["role"]]

    department = metadata.get("department", "general")

    required_level = DOC_ACCESS[department]

    if user_level < required_level:
        return False

    if department not in user["departments"] and department != "general":
        return False

    return True
```

```python
def retrieve_documents(query, username, k=5):

    docs = vector_db.similarity_search(query, k=k)

    authorized_docs = []

    for doc in docs:
        if is_authorized(username, doc.metadata):
            authorized_docs.append(doc)

    return authorized_docs
```

```python
query = "finance budget risks"

results = retrieve_documents(query, "alice")

for doc in results:
    print("SOURCE:", doc.metadata["source"])
    print(doc.page_content)
    print("-" * 50)
```

```python
SYSTEM_PROMPT = """
You are a secure enterprise AI assistant.

Answer ONLY from the provided context.

Do not hallucinate.

Provide citations.
"""
```

```python
def generate_answer(query, docs):

    context = "\n\n".join([
        d.page_content for d in docs
    ])

    citations = list(set([
        d.metadata["source"] for d in docs
    ]))

    prompt = f"""
Context:
{context}

Question:
{query}
"""

    response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[
            {
                "role": "system",
                "content": SYSTEM_PROMPT
            },
            {
                "role": "user",
                "content": prompt
            }
        ],
        temperature=0.1
    )

    answer = response.choices[0].message.content

    return {
        "answer": answer,
        "citations": citations,
        "confidence": "High" if len(docs) >= 2 else "Medium"
    }
```

```python
query = "What finance risks were detected?"

retrieved_docs = retrieve_documents(query, "alice")

response = generate_answer(query, retrieved_docs)

response
```
